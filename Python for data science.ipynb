```python
m=np.matrix("1,2,3,4;4,5,6,7;4,3,2,3")
print(m.shape)
print(m.size)
m2=np.insert(m,1,[[2,3,4,5]],axis=0)
m1=np.delete(m2,3,1)
print('m1=',m1,'\nm2=',m2)
print('m2*m1=',m2*m1)                # Matrixx multiplication
print(np.multiply(m.transpose(),m1)) # Elementwise multiplication
np.dot(m2,m1)                        # For matrices it will do matrix multiplication
```

### Linalg Module in numpy

```python
import numpy.linalg as lp
print('|m2|=',lp.det(m2))
print('rank(m2)=',lp.matrix_rank(m2))
try:
    print('m2^-1=',lp.inv(m2))
except:
    print('m2 is singular matrix')
else:
    print('m2 is non singular matrix')
m=np.matrix("1,2,3;4,5,6;3,2,6")
print('m^-1=',lp.inv(m))
b=np.matrix("1,2,3")
print(lp.solve(m,b.transpose()))
```

### Shalow copy vs Deep copy

- Some times we don't want to alter the master dataframe and want to work on a specific part of the data frame or want to make some temporary changes for the analysis.
- As dataframe is mutable, simple copy using assignment statement won't work.
- So to make a deep copy of it we will use `copy` module of the data frame.

```python
df=df4.copy(deep=True)
#Some common commands
print(df.size,df.shape,df.ndim,sep='\n')# Notice that methods are not ending with ()
print(df.index,df.columns,sep='\n')
print(df.memory_usage())
print(df.head(6))
print(df.tail(5))
```

## Dealing with missing values

```python
df.describe()
```

```python
df=pd.read_csv('Toyota.csv',index_col=0,na_values=['??','????'])
df.Age.mean()
df.Age.fillna(df.Age.mean(),inplace=True)
df.HP.median()
df.HP.fillna(df.HP.median(),inplace=True)
df.isnull().sum()
df.FuelType.value_counts()
df.FuelType.fillna(df.FuelType.value_counts().index[0],inplace=True)
df.MetColor.mode()
df.MetColor.fillna(df.MetColor.mode()[0],inplace=True) #Don't forget to add zero here
df.KM.fillna(df.KM.mean(),inplace=True)
df.isnull().sum()
```

### Filling values at one stretch

```python
df=pd.read_csv('Toyota.csv',index_col=0,na_values=['??','????'])
df.isnull().sum()
df=df.apply(lambda x:x.fillna(x.mean()) if x.dtype=='float' else x.fillna(x.value_counts().index[0]))
df.isnull().sum()
```
