```python
#Necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os,sys
os.chdir(r'E:\Sem6\PRML\PA_2\PA2_Datasets\PA2_Datasets')
```

```python
df = pd.read_csv('Regression_dataset.csv')
```

```python
X = np.array(df.iloc[:,:13])
y = np.array(df.iloc[:,13])
df.shape
```

```python
print(X.shape,y.shape)
```

```python
#For splitting the data
from sklearn.model_selection import train_test_split as tts
```

```python
X_train,X_test,y_train,y_test=tts(X,y,test_size=0.20,random_state=42)
```

```python
#Training kernel ridge regression with linear kernel-No kernel ridge regression
from sklearn.kernel_ridge import KernelRidge
reg = KernelRidge(alpha=1.0,kernel='linear')
reg.fit(X_train,y_train)
y_pred=reg.predict(X_test)
```

```python
#MSE for training and test data sets
from sklearn.metrics import mean_squared_error
test_error=mean_squared_error(y_test,y_pred)
training_error = mean_squared_error(y_train,reg.predict(X_train))
```

```python
plt.scatter(reg.predict(X_train),y_train,label='data points')
plt.plot(np.arange(0,60,0.1),np.arange(0,60,0.1),'r-',label='y=x line')
plt.xlabel('Predicted y')
plt.ylabel('True y')
plt.title('Scatter plot for results of train data')
plt.legend()
```

```python
plt.scatter(y_pred,y_test,label='data points')
plt.plot(np.arange(0,60,0.1),np.arange(0,60,0.1),'r-',label='y=x line')
plt.xlabel('Predicted y')
plt.ylabel('True y')
plt.title('Scatter plot for results of test data')
plt.legend()
```

```python
#Training kernel ridge regression with polynomial regression
reg2=KernelRidge(alpha=1.0,kernel='polynomial',degree=2,coef0=1)
reg2.fit(X_train,y_train)
y_pred2=reg2.predict(X_test)
```

```python
#MSE error for polynomial kernel ridge regression
test_error2=mean_squared_error(y_test,y_pred2)
training_error2 = mean_squared_error(y_train,reg2.predict(X_train))
```
